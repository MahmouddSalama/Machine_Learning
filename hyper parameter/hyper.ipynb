```python
from sklearn.datasets import load_iris
iris=load_iris()
```

```python
import pandas as pd
import numpy as np

data=pd.DataFrame(iris.data,columns=iris.feature_names)
data.head()
```

```python
data['target']=iris.target
data.head()
```

## 1 Approch in Train test split

```python
from sklearn.model_selection import train_test_split

Xtrain,Xtest,Ytrain,Ytest=train_test_split(data.drop(['target'],axis=1),data['target'],train_size=.8)
```

```python
from sklearn.svm import SVC
svc=SVC(C=10,kernel='rbf')
svc.fit(Xtrain,Ytrain)
svc.score(Xtest,Ytest)
```

## 2  K fold Cross Validation

```python
X=data.drop(['target'],axis=1)
Y=data['target']
```

```python
from sklearn.model_selection import cross_val_score

c =cross_val_score(SVC(C=10,kernel='rbf',gamma='auto'),X,Y,cv=5)
c
```

```python
c =cross_val_score(SVC(C=20,kernel='linear',gamma='auto'),X,Y,cv=5)
c
```

```python
c =cross_val_score(SVC(C=1,kernel='linear',gamma='auto'),X,Y,cv=5)
c
```

```python
kernel=['linear', 'poly', 'rbf',]
c=[1,10,20,30]
avg_scores={}
for k in kernel:
    for cc in c:
        cvScore= cross_val_score(SVC(C=cc,kernel=k,gamma='auto'),X,Y,cv=5)
        avg_scores[k + '_' + str(cc)] = np.average(cvScore)

avg_scores
```

## 3 use   GridsearchCV

```python
from sklearn.model_selection import GridSearchCV

clf=GridSearchCV(SVC(gamma='auto'),{
    'C':[1,10,20,30],
    'kernel':['linear', 'poly', 'rbf',]
}, cv=5,return_train_score=False
)
clf.fit(X,Y)
clf.cv_results_
d=pd.DataFrame(clf.cv_results_)
d
```

```python
clf.best_params_
```

```python
pd.DataFrame(clf.cv_results_)[['param_C','param_kernel','mean_test_score']]
```

# Main Example 

```python
# import models
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
```

```python
# dic of models and params

model_params={
    'SVC':{
        'model':SVC(gamma='auto'),
        'params':{
            'C':[1,10,20,30],
            'kernel':['linear', 'poly', 'rbf',]
        }
    },
    'RandomForestClassifier':{
        'model':RandomForestClassifier(),
        'params':{
            'n_estimators':[5,10,15,20,30,40,50,],
            'criterion':["gini", "entropy"]
        }
    },
    "LogisticRegression":{
        'model':LogisticRegression(),
        'params':{
            'C':[1,5,10,15]
        }
    }
}
```

```python
scores=[]
for mn,mp in model_params.items():
    clf=GridSearchCV(mp['model'],mp['params'], cv=5,return_train_score=False)
    clf.fit(X,Y)
    scores.append({
        'modelName':mn,
        'bestParams':clf.best_params_,
        'bestScore':clf.best_score_
    })

scores
```

```python
df = pd.DataFrame(scores)
df
```
