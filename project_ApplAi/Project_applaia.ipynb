```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
data=pd.read_csv('Applais project.csv')
data.head()
```

```python
data['mobile_wt'].fillna(data['mobile_wt'].mean(),inplace=True)
```

```python
plt.figure(figsize=(25,20))
sns.set_context('paper',font_scale=1.4)
sns.heatmap(data.corr(),annot=True,cmap='Blues')
```

```python
plt.figure(figsize=(15,10))
sns.countplot(data=data,x='fc',hue='four_g')
```

```python
from sklearn.cluster import KMeans
```

```python
sse=[]
k_rng=range(1,10)
for k in k_rng:
    km=KMeans(n_clusters=k)
    km.fit(data)
    sse.append(km.inertia_)
```

```python
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_rng,sse)
```

```python
km=KMeans(n_clusters=3)
y_predict=km.fit_predict(data)
y_predict
```

```python
data['Cluster']=y_predict
data.head()
```

```python
plt.figure(figsize=(25,20))
sns.set_context('paper',font_scale=1.4)
sns.heatmap(data.corr(),annot=True,cmap='Blues')
```

```python
plt.figure(figsize=(15,10))
sns.countplot(data=data,x='ram',hue='Cluster')
```

```python
sub_data=data[['battery_power','clock_speed','four_g','int_memory','mobile_wt','n_cores','ram','three_g','wifi','Cluster']]
tareget=data['Cluster']
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(sub_data,tareget,train_size=0.8)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model2=LogisticRegression()
model2.fit(x_train,y_train)
model2.score(x_test,y_test)
```
